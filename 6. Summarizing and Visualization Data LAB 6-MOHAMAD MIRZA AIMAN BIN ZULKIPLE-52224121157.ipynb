```python
## IMPORT PANDAS
import pandas as pd
```

```python
import numpy as np
```

```python

xls = pd.ExcelFile('yelp.xlsx')
df= xls.parse('yelp_data')
```

```python
## to see all column in dataframe for file yelp.xlsx
df.head()
```

Joining Data

```python
## Join data cities to dataframe
df_cities = xls.parse('cities')
```

```python
## to see the availability of columns city into dataframe 
df_cities.head()
```

```python
df=pd.merge(left=df, right=df_cities,  how='inner', left_on='city_id',  right_on='id')
```

```python
df.head()
```

```python
## to add columns states into dataframe
df_states = xls.parse('states')
```

```python
## to see the availability of columns states into dataframe 
df_states.head()
```

```python
df=pd.merge(left=df, right=df_states, how='inner',  left_on='state_id',  right_on='id')
```

```python
df.shape
```

```python
df.head()
```

```python
### To delete columns id_x
del df['id_x']
```

```python
### To delete columns id_y
del df['id_y']
```

```python
df.head()
```

Updating and Creating Data

```python
df["rating"] = df["stars"] *2
```

```python
def convert_to_rating(x):
    return(str(x) + " out of 10")
```

```python
df["rating"] = df["rating"].apply(convert_to_rating)
```

```python
df.head()
```

Histograms

```python
import matplotlib.pyplot as plt
```

```python
%pylab inline
```

```python
## to filter dataframe city of Pittsburgh
df_pitt = df[df["city"] =="Pittsburgh"]
```

```python
df_pitt.head()
```

```python
df_vegas = df[df["city"] == "Las Vegas"]
```

```python
df_vegas.head()
```

```python
pitt_stars = df_pitt["stars"]
vegas_stars = df_vegas["stars"]
vegas_stars.head()
```

```python
plt.hist(
    pitt_stars,
    alpha = 0.3,
    color='blue',
    label ='Pittsburgh',
    bins = 'auto'
)

plt.hist (
   vegas_stars,
    alpha = 0.3,
    color='purple',
    label='Las Vegas',
    bins='auto'
)    

plt.xlabel("Rating")
plt.ylabel("Number of Rating Scores")

plt.legend(loc = 'best')

plt.title("Review distribution of Pittsburgh and Las Vegas")

plt.show()
```

```python
plt.hist (
   [pitt_stars,vegas_stars],
    alpha = 0.7,
    color=['green', 'orange'],
    label= ['Pittsburgh','Las Vegas'],
    bins='auto'
)

plt.xlabel('Rating')
plt.ylabel('Number of Rating Scores')

plt.legend(loc = 'best')

plt.title('Review distribution of Pittsburgh and Las Vegas')

plt.show()
```

Scatterplots

```python
df_health= df[df["category_0"] == "Health & Medical"]
df_fast= df[df["category_0"] == "Fast Food"]
df_break= df[df["category_0"] == "Breakfast & Brunch"]
df_break.head()
```

```python
df_fast.head()
```

```python
df_health.head()
```

```python
plt.scatter(
    df_health["stars"], df_health["review_count"],
    marker = "o",
    color = 'r',
    alpha = 0.7,
    s = 124,
    label = ['Health & Medical']
)

plt.scatter(
    df_fast["stars"], df_fast["review_count"],
    marker = "h",
    color = 'b',
    alpha = 0.7,
    s = 124,
    label = ['Fast Food']
)

plt.scatter(
    df_break["stars"], df_break["review_count"],
    marker = "^",
    color = 'g',
    alpha = 0.7,
    s = 124,
    label = ['Breakfast & Brunch']
)

plt.xlabel('Rating')
plt.ylabel('Review Count')

plt.legend(loc = 'upper left')

axes = plt.gca()
axes.set_yscale('log')

plt.show() 
```
